```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error

path = './comparison_target_pred_fold10/fold1_comparison.csv'

df = pd.read_csv(path)

# 원본

df['y_pred'].tolist()
df['y_target'].tolist()


df_ypred = df['y_pred'].tolist()
df_target = df['y_target'].tolist()
plt.title(f'50sample pred value original MAE : {round(mean_absolute_error(df_target[:50], df_ypred[:50]),4)}')
plt.plot(df_ypred[:50], label='y_pred', alpha = 0.4)
plt.plot(df_target[:50], label='y_target', alpha = 0.4)
plt.xlabel('sample number')
plt.ylabel('target value')
plt.legend()
plt.show()


print(f'total dataset original MAE : {round(mean_absolute_error(df_target, df_ypred),4)}')
```

```python
# 반올림

df_ypred = df['y_pred'].round().tolist()
df_target = df['y_target'].tolist()

plt.title(f'50sample pred value round MAE : {round(mean_absolute_error(df_target[:50], df_ypred[:50]),4)}')
plt.plot(df_ypred[:50], label='y_pred', alpha = 0.4)
plt.plot(df_target[:50], label='y_target', alpha = 0.4)
plt.xlabel('sample number')
plt.ylabel('target value')
plt.legend()
plt.show()

print(f'total dataset round MAE : {round(mean_absolute_error(df_target, df_ypred),4)}')
```

```python
# 내림

df_ypred = df['y_pred'].apply(np.floor).tolist()
df_target = df['y_target'].tolist()

plt.title(f'50sample pred value round down MAE : {round(mean_absolute_error(df_target[:50], df_ypred[:50]),4)}')
plt.plot(df_ypred[:50], label='y_pred', alpha = 0.4)
plt.plot(df_target[:50], label='y_target', alpha = 0.4)
plt.xlabel('sample number')
plt.ylabel('target value')
plt.legend()
plt.show()

print(f'total dataset round down MAE : {round(mean_absolute_error(df_target, df_ypred),4)}')
```

```python
# 올림

df_ypred = df['y_pred'].apply(np.ceil).tolist()
df_target = df['y_target'].tolist()

plt.title(f'50sample pred value round up MAE : {round(mean_absolute_error(df_target[:50], df_ypred[:50]),4)}')
plt.plot(df_ypred[:50], label='y_pred', alpha = 0.4)
plt.plot(df_target[:50], label='y_target', alpha = 0.4)
plt.xlabel('sample number')
plt.ylabel('target value')
plt.legend()
plt.show()

print(f'total dataset round up MAE : {round(mean_absolute_error(df_target, df_ypred),4)}')
```
