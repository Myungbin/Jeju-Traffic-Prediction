```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error

for i in range(1,11):
    df = pd.read_csv(f'./comparison_target_pred_fold10/fold{i}_comparison.csv')

    y_pred = df['y_pred']
    y_target = df['y_target']

    mae = round(mean_absolute_error(y_target, y_pred),4)
    sns.kdeplot(y_pred, shade=True, label='y_pred')
    sns.kdeplot(y_target, shade=True, label='y_target')
    plt.xlabel('target value')
    plt.title(f'FOLD{i} MAE:{mae} KDE plot')
    plt.legend()
    plt.show()

```

Fold별로 MAE값의 차이가 나며 타겟 속력이 20~40km/h 50~60km/h의 예측에서 값의 density가 차이 남을 확인할 수 있다.

-> 예측값이 20~40km/h, 50~60km/h 일 경우 타 구간에 비해 변동성이 조금 더 크게 나타난다.
