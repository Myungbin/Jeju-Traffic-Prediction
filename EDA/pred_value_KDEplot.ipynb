```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error

for i in range(1,11):
    df = pd.read_csv(f'./comparison_target_pred_fold10/fold{i}_comparison.csv')

    y_pred = df['y_pred']
    y_target = df['y_target']

    mae = round(mean_absolute_error(y_target, y_pred),4)
    sns.kdeplot(y_pred, shade=True, label='y_pred')
    sns.kdeplot(y_target, shade=True, label='y_target')
    plt.xlabel('target value')
    plt.title(f'FOLD{i} MAE:{mae} KDE plot')
    plt.legend()
    plt.show()

```
